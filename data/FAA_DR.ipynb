# Did DataRobot find new spy planes that Buzzfeed did not?

```python
preds = pd.read_csv('DR_results_ExtraTrees.csv')
preds.shape
```

```python
buzzfeed_pics = pd.read_csv('candidates.csv')
buzzfeed_pics.shape
```

```python
buzzfeed_pics.tail()
```

```python
preds.sort_values(by='Prediction').tail(100)
```

```python
# when I ran predictions in DataRobot, I forgot to include other columns of interest, like adshex, squawk, etc.  Read in
#prediction.csv which has our good planes, and do an inner join with my predictions data
good_planes = pd.read_csv('predict.csv')
good_planes.head(10)
```

```python
good_planes.drop('Unnamed: 0', inplace=True, axis=1)
```

```python
#make the index the row_id
good_planes.reset_index(inplace=True)
good_planes.head()
```

```python
good_planes.rename(columns={'index':'row_id'}, inplace=True)
```

```python
#spot check row index is the right merge
good_planes[4070:4071]
```

```python
good_planes.drop('level_0', inplace=True, axis=1)
```

```python
good_planes[2169:2170]
```

```python
good_planes.columns
```

```python
#merge files on row_id
planes_wpreds = preds.merge(good_planes)
```

```python
planes_wpreds.tail()
```

```python
#are our top 72 planes in their top 100?planes
len((set(planes_wpreds.adshex[planes_wpreds.spyplane>0]).difference(buzzfeed_pics.adshex[0:72]))) #in planes but excluding fed
```

```python
dr_top_100=planes_wpreds.sort_values(by='Prediction', ascending=True).tail(100)
dr_top_100
```

```python
#are our top 100 planes the same as theirs?
dr_dif_pics = ((set(dr_top_100.adshex).difference(buzzfeed_pics.adshex))) #in dr_top_100 but excluding buzz_feed_top 100
len(dr_dif_pics)
```

```python
#did we predict on training data?
print(len((set(train.adshex).difference(planes_wpreds.adshex))))
print(planes_wpreds[planes_wpreds.adshex.isin(train.adshex)])
```

```python
dr_top_100.adshex.sort_values()[0:15]
```

```python
buzzfeed_pics.adshex.sort_values()[0:15]
```

```python
buzzfeed_pics.tail(15)
```

```python
#our top 15 aren't the same
dr_top_100[dr_top_100.adshex.isin(dr_dif_pics)].sort_values(by='Prediction', ascending=False)
```

```python
#looks like we found some high probability pics that didn't make it into buzzfeeds algorithm.  Our top one has a predicted probability
#of being a spyplane of 0.65.  Let's see how many buzzfeed planes have a probability higher than 0.65
len(buzzfeed_pics[buzzfeed_pics.surveil>0.60])
```
